```python
import numpy as np
import matplotlib.pyplot as pl
import csv

data_path = 'data1.csv'
with open(data_path, 'r') as f:
    reader = csv.reader(f, delimiter=',')
    headers = next(reader)
    data = np.array(list(reader)).astype(float)

pl.figure(1,figsize=(10,8))
pl.plot(data,'.-')
pl.grid(True)    
```
