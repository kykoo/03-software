```python
import datetime
from numpy import *

nfiles = len(files)

t0 = datetime.datetime(2020,5,8,0,0,0).timestamp()
t1 = datetime.datetime(2020,5,8,6,0,0).timestamp()
ntarget = (t1-t0)//120
count = 0
for t in arange(t0,t1,60*2):
    filename_ = datetime.datetime.fromtimestamp(t).strftime('acc2-%Y-%m%d-%H%M%S')
    found = False
    for file in files:
        if filename_ in file:
            found = True
            break
    if not found:
        print('{} not found'.format(filename_))
        count += 1

print('{:.0f}%'.format(count/ntarget*100))        
```
