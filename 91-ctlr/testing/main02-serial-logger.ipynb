```python
# Connection 
import serial
import time
import sys

ser =  serial.Serial('/dev/ttyACM0', 115200, timeout=1) 
line = ser.readline()   # read a '\n' terminated line
print(line)
ser.close()
```

```python
# Connection when disconnected
import serial
import time
import sys

try:
    ser =  serial.Serial('/dev/ttyACM0', 115200, timeout=1) 
    line = ser.readline()   # read a '\n' terminated line
    print(line)
except serial.SerialException as exc:
    print('Exception detected')
    if exc.errno==2:
        print('could not open port')
    
ser.close()
```

```python
# Disconnect Exception
import serial
import time
import sys

try:
    ser =  serial.Serial('/dev/ttyACM0', 115200, timeout=1) 
    line = ser.readline()   # read a '\n' terminated line
    print(line)
except serial.SerialException as exc:
    print('Exception detected')
    sys.exit()
    if exc.errno==2:
        print('could not open port')

while True:
    try:
        line = ser.readline()   # read a '\n' terminated line
        print(line)
    except serial.SerialException as exc:
        ser.close()
        print('errno={}'.format(exc.errno))
        raise
            
ser.close()
```

```python
# Logger
import argparse
import serial
import time
import sys
import datetime
import path
import os

parser = argparse.ArgumentParser(description='log serial outputs')
parser.add_argument('serialPortNumber', help='serial port number /dev/ttyACM{i}')
args = parser.parse_args()

serialPort = '/dev/ttyACM{}'.format(args.serialPortNumber)

filename = 'ctlr-{}.log'.format(datetime.datetime.now().strftime('%Y%m%d-%H%M00'))
serialPortNum = 0
serialPort = '/dev/ttyACM{}'.format(serialPortNum)

with open(filename,'a') as fp:
    while True:
        # Check port exist
        print('****checking if {} exists.',end='')
        while True:
            if os.path.exists(serialPort):
                break
            print('.', end='')
            time.sleep(1)
            
        # Opening port
        while True:
            try:
                ser =  serial.Serial('/dev/ttyACM0', 115200, timeout=1) 
                print('')
                break
            except serial.SerialException as exc:
                if exc.errno==2:
                    print('.',end='')
                    time.sleep(1)
        # Read and Write
        try:
            while True:
                line = ser.readline()   # read a '\n' terminated line
                line_asc=line.decode('ASCII')
                print(line_asc,end='')
                fp.write(line_asc)
                fp.flush()
        except serial.SerialException as exc:
            ser.close()
            print('serial error detected. closing')

            
```
