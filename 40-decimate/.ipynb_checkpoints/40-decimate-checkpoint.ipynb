# Decimate by scipy.signal

```python
from numpy import *
from scipy.signal import *
from matplotlib.pyplot import *


def readAcc(fileName):
    t = arange(0,180,0.01)
    a = []
    with open(fileName, 'r') as file:
        lines = file.readlines()
        for line in lines:
            vals = line.split(',')
            vals = [float(vals[0]), float(vals[1]), float(vals[2])]
            a.append(vals)
    return t, array(a)

fileName = 'acc-80.txt'
Fs = 100
t, acc = readAcc(fileName)
# f, Pxx = welch(a, Fs, nperseg=2**10, axis=0)
# figure(1, figsize=(8,6))
# semilogy(f,Pxx)
# xlim([0, Fs/2])
# xlabel('Frequency (Hz)')
# ylabel('PSD')
# grid(True)
# title(fileName)
# show()
acc_mean = mean(acc,axis=0)
acc_det = acc - acc_mean

# plot(a[:,0:2])
# plot(a[:,0:2] - a_mean[0:2])

acc_det_d = decimate(acc_det,4,axis=0)
t_d = t[0:-1:4]

acc_d = acc_det_d + acc_mean

if 1:
    figure(1)
    clf()
    subplot(221)
    plot(t,acc[:,0])
    plot(t_d,acc_d[:,0],'.-')
    xlim([0,180])
    grid(True)
    subplot(223)    
    plot(t,acc[:,0])
    plot(t_d,acc_d[:,0],'.-')
    xlim([0,1])
    grid(True)
    subplot(122)
    f, Pxx = welch(acc, Fs, nperseg=2**10, axis=0)
    f_d, Pxx_d = welch(acc_d, Fs/4, nperseg=2**8, axis=0)
    plot(f,20*log10(Pxx[:,2]))
    plot(f_d,20*log10(Pxx_d[:,2]))
    xlim([0,20])
    ylim([-140,-50])


```

# My implementation

```python
# Filter design

n_filter = 8
q = 4
b, a = cheby1(n_filter, 0.05, 0.8 / q)
print('a={}'.format(a))
print('b={}'.format(b))

# Implementation of lfilter
#    Based on https://github.com/scipy/scipy/blob/master/scipy/signal/signaltools.py
# 
# The filter function is implemented as a direct II transposed structure.
#     This means that the filter implements::
#        a[0]*y[n] = b[0]*x[n] + b[1]*x[n-1] + ... + b[M]*x[n-M]
#                              - a[1]*y[n-1] - ... - a[N]*y[n-N]
#     where `M` is the degree of the numerator, `N` is the degree of the
#     denominator, and `n` is the sample number.  It is implemented using
#     the following difference equations (assuming M = N)::
#          a[0]*y[n] = b[0] * x[n]               + d[0][n-1]
#            d[0][n] = b[1] * x[n] - a[1] * y[n] + d[1][n-1]
#            d[1][n] = b[2] * x[n] - a[2] * y[n] + d[2][n-1]
#          ...
#          d[N-2][n] = b[N-1]*x[n] - a[N-1]*y[n] + d[N-1][n-1]
#          d[N-1][n] = b[N] * x[n] - a[N] * y[n]
#     where `d` are the state variables.

m = 3
d_prev = zeros((n_filter,m))
d_now = zeros((n_filter,m))
N = acc.shape[0]
x = acc
x_mean = mean(x,axis=0)
x = x - x_mean
y = zeros((N,m))
for i in range(N):
    y[i,:] = b[0]*x[i,:] + d_prev[0,:]
    for j in range(n_filter):
        d_now[j,:]= b[j+1]*x[i,:] - a[j+1]*y[i,:] + d_prev[j+1,:] if j < n_filter-1 else \
                    b[j+1]*x[i,:] - a[j+1]*y[i,:] 
    d_prev = d_now
y = y + x_mean

# Visualisation

figure(1)
plot(t_d,y[0:-1:4,0],'ro')
plot(t_d,acc_d[:,0],'g.')
xlim([0,5])
grid(True)

f, Pxx = welch(acc, Fs, nperseg=2**10, axis=0)
Fd,Pxxd = welch(y[0:-1:4,:],Fs/4,window='hann',nperseg=2**8,axis=0)

figure(2)
plot(f,20*log10(Pxx[:,0]))
plot(Fd,20*log10(Pxxd[:,0]))
xlim([0,20])
ylim([-180,-50])
```
